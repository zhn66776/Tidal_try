```python
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
import cartopy.feature as cfeature

locations = {
    'Aberdeen': (-2.08013, 57.14406),
    'Bournemouth': (-1.87486, 50.71433),
    'Tobermory': (-6.06412, 56.62311),
    'Millport': (-4.90583, 55.74964)
}

fig, ax = plt.subplots(figsize=(12, 8), subplot_kw={'projection': ccrs.PlateCarree()})

extent = [-8, 2, 50, 58]
ax.set_extent(extent, crs=ccrs.PlateCarree())

ax.add_feature(cfeature.BORDERS, linestyle=':')
ax.add_feature(cfeature.LAND, edgecolor='gray', facecolor='lightgray')
ax.add_feature(cfeature.OCEAN, edgecolor='black', facecolor='lightblue')

coast_feature = cfeature.NaturalEarthFeature(
    category='physical', name='coastline', scale='10m',
    edgecolor='darkgray',
    facecolor='none',
    linewidth=1
)
ax.add_feature(coast_feature)

ax.set_xticks([-8, -6, -4, -2, 0, 2], crs=ccrs.PlateCarree())
ax.set_yticks([50, 52, 54, 56, 58], crs=ccrs.PlateCarree())

ax.set_xticklabels(
    ['-8°', '-6°', '-4°', '-2°', '0°', '2°'],
    fontsize=12, fontname='Times New Roman'
)
ax.set_yticklabels(
    ['50°', '52°', '54°', '56°', '58°'],
    fontsize=12, fontname='Times New Roman'
)

ax.set_xlabel('Longitude', fontsize=14, fontname='Times New Roman')
ax.set_ylabel('Latitude', fontsize=14, fontname='Times New Roman')

for name, (lon, lat) in locations.items():
    ax.plot(lon, lat, 'ro', transform=ccrs.PlateCarree())
    ax.text(
        lon + 0.2, lat + 0.15, name,
        transform=ccrs.PlateCarree(),
        fontsize=16,
        fontname='Times New Roman',
        color='black'
    )
plt.show()
```
